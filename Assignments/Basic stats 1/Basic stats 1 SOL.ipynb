```python
import matplotlib.pyplot as plt

# Bar graph for date
df.groupby('Date')['Total Sales Value'].sum().plot(kind='bar')
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Sales by Date")
plt.show()

# Bar graph for day
df.groupby('Day')['Total Sales Value'].sum().plot(kind='bar')
plt.xlabel("Day")
plt.ylabel("Sales")
plt.title("Sales by Day")
plt.show()

# Bar graph for sku
df.groupby('SKU')['Total Sales Value'].sum().plot(kind='bar')
plt.xlabel("SKU")
plt.ylabel("Sales")
plt.title("Sales by SKU")
plt.show()

# Bar graph for city
df.groupby('City')['Total Sales Value'].sum().plot(kind='bar')
plt.xlabel("City")
plt.ylabel("Sales")
plt.title("Sales by City")
plt.show()

# Bar graph for bu
df.groupby('BU')['Total Sales Value'].sum().plot(kind='bar')
plt.xlabel("BU")
plt.ylabel("Sales")
plt.title("Sales by BU")
plt.show()

# Bar graph for brand
df.groupby('Brand')['Total Sales Value'].sum().plot(kind='bar')
plt.xlabel("Brand")
plt.ylabel("Sales")
plt.title("Sales by Brand")
plt.show()

# Bar graph for model
df.groupby('Model')['Total Sales Value'].sum().plot(kind='bar')
plt.xlabel("Model")
plt.ylabel("Sales")
plt.title("Sales by Model")
plt.show()
```
