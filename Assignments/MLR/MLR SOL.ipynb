```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
df['Fuel_Type'] = LE.fit_transform(df['Fuel_Type'])
df.head()
```

```python
# standadization
df.iloc[:, 0:].head()
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
df.iloc[:, 0:] = SS.fit_transform(df.iloc[:, 0:])
df.iloc[:, 0:].head()
```

```python
df.corr()
```

```python
Y=df["Price"]
X=df.iloc[:,1:]
```

```python
X
```

```python
Y
```

```python
# Data partition
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# Model Fitting
from sklearn.linear_model import LinearRegression
LR= LinearRegression()
```

```python
LR.fit(X_train,Y_train)
```

```python
print(LR.coef_)
```

```python
LR2=LinearRegression()
LR2.fit(X_train[['Age_08_04', 'KM', 'HP', 'cc', 'Doors', 'Gears', 'Weight']], Y_train)
print(LR2.coef_)
```

```python
LR3=LinearRegression()
LR3.fit(X_train[['Age_08_04', 'Weight', 'KM', 'HP','Doors','cc', 'Gears', 'Automatic' ]], Y_train)
print(LR3.coef_)
```

```python
# Predictions
Y_pred_train = LR.predict(X_train)
Y_pred_test = LR.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mse1 = mean_squared_error(Y_train, Y_pred_train)
mse2 = mean_squared_error(Y_test, Y_pred_test)

mae1 = mean_absolute_error(Y_train, Y_pred_train)
mae2 = mean_absolute_error(Y_test, Y_pred_test)

r2_1 = r2_score(Y_train, Y_pred_train)
r2_2 = r2_score(Y_test, Y_pred_test)
```

```python
print("Train_Mean squared error:", mse1)
print("Test_Mean squared error:", mse2)

print("Train_Mean absolute error:", mae1)
print("Test_Mean absolute error:", mae2)

print("Train_R-squared score:", r2_1)
print("Test_R-squared score:", r2_2)
```

```python
# Predictions for model 2 
Y_pred2 = LR2.predict(X_test[['Age_08_04', 'KM', 'HP', 'cc', 'Doors', 'Gears', 'Weight']])
mse2 = mean_squared_error(Y_test,Y_pred2)
mae2 = mean_absolute_error(Y_test, Y_pred2)
r2_2 = r2_score(Y_test, Y_pred2)

print("Model 2:")
print("Mean squared error:", mse2)
print("Mean absolute error:", mae2)
print("R-squared score:", r2_2)
```

```python
# Predictions for model 3 
Y_pred3= LR3.predict(X_test[['Age_08_04', 'Weight', 'KM', 'HP','Doors','cc', 'Gears', 'Automatic' ]])
mse3 = mean_squared_error(Y_test,Y_pred3)
mae3 = mean_absolute_error(Y_test, Y_pred3)
r2_3 = r2_score(Y_test, Y_pred3)

print("Model 3:")
print("Mean squared error:", mse3)
print("Mean absolute error:", mae3)
print("R-squared score:", r2_3)
```

```python
# LASSO AND RIDGE REGRESSION
from sklearn.linear_model import Lasso, Ridge

lasso_model = Lasso(alpha=1.0)
lasso_model.fit(X_train, Y_train)
Y_pred_lasso = lasso_model.predict(X_test)
mse_lasso = mean_squared_error(Y_test, Y_pred_lasso)
mae_lasso = mean_absolute_error(Y_test, Y_pred_lasso)
r2_lasso = r2_score(Y_test, Y_pred_lasso)

ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train, Y_train)
Y_pred_ridge = ridge_model.predict(X_test)
mse_ridge = mean_squared_error(Y_test, Y_pred_ridge)
mae_ridge = mean_absolute_error(Y_test, Y_pred_ridge)
r2_ridge = r2_score(Y_test, Y_pred_ridge)

print("Lasso Model:")
print("Mean squared error:", mse_lasso)
print("Mean absolute error:", mae_lasso)
print("R-squared score:", r2_lasso)

print("Ridge Model:")
print("Mean squared error:", mse_ridge)
print("Mean absolute error:", mae_ridge)
print("R-squared score:", r2_ridge)
```

```python
How They Are Helpful:

1.Improved Model Performance: Normalization and standardization can improve the performance and convergence of machine learning algorithms, especially those sensitive to the scale of the input features, such as gradient descent-based algorithms.
2.Equal Treatment of Features: By bringing all features to a similar scale, normalization and standardization ensure that no single feature dominates the learning process due to its larger magnitude.
3.Enhanced Interpretability: Standardization, in particular, can improve the interpretability of coefficients or feature importance in models like linear regression by ensuring that coefficients represent the change in the target variable for a standard deviation change in the feature.
4.Robustness: Normalization and standardization are robust techniques that are less affected by outliers compared to other scaling methods.
In summary, normalization and standardization are essential preprocessing steps that help to prepare data for modeling, improve algorithm performance, and ensure fair treatment of features. They are especially crucial in scenarios where features have different scales or units.
```
