```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(one_hot_encoded, min_support=0.05, use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
print("Frequent Itemsets:")
print(frequent_itemsets)
```

```python
print("\nAssociation Rules:")
print(rules)
```

```python
# Apply the Apriori algorithm with a lower support threshold
frequent_itemsets = apriori(one_hot_encoded, min_support=0.2, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Display the frequent itemsets
print("Frequent Itemsets:")
print(frequent_itemsets)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
```

```python
print("\nAssociation Rules:")
print(rules)
```

```python
rules['lift'] = pd.to_numeric(rules['lift'])

# Display the top 10 rules with the highest lift
top_rules_lift = rules.nlargest(10, 'lift')
print("\nTop 10 rules with the highest lift:")
print(top_rules_lift)
```

```python
rules['confidence'] = pd.to_numeric(rules['confidence'])

# Display the top 10 rules with the highest confidence
top_rules_confidence = rules.nlargest(10, 'confidence')
print("\nTop 10 rules with the highest confidence:")
print(top_rules_confidence)
```

```python
high_confidence_rules = rules[(rules['lift'] > 1) & (rules['confidence'] > 0.7)]
print("\nRules with lift > 1 and confidence > 0.7:")
print(high_confidence_rules)
```

```python
#The output indicates that there are no rules in the DataFrame rules that meet the criteria of having the top 10 highest lift values. This means that all the lift values in the DataFrame are either missing or equal to zero.

#The output indicates that there are no rules in the DataFrame rules that meet the criteria of having the top 10 highest confidence values.This means that all the confidence values in the DataFrame are below the confidence threshold

#The output indicates that there are no association rules in the DataFrame rules that meet the criteria of having both a lift greater than 1 and a confidence greater than 0.7.This means that none of the rules in the dataset satisfy these conditions simultaneously.
```
