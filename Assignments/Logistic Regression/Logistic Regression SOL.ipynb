```python
def avg_age(i):
    Age = i[0]
    Pclass = i[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 38
        elif Pclass == 2:
            return 29
        else:
            return 25
    else:
        return Age
```

```python
train_df['Age'] = train_df[['Age','Pclass']].apply(avg_age,axis=1)
test_df['Age'] = test_df[['Age','Pclass']].apply(avg_age,axis=1)
```

```python
sex=pd.get_dummies(train_df["Sex"],drop_first=True)

Embarked=pd.get_dummies(train_df["Embarked"],drop_first=True)

train_df.drop(["Name","Sex","Ticket","Embarked"],axis=1,inplace=True)

train_df=pd.concat([train_df,sex,Embarked],axis=1)
```

```python
train_df.head()
```

```python
col=["male","Q","S"]
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in col:
    train_df[i]=le.fit_transform(train_df[i])
train_df.head()
```

```python
x=train_df.drop("Survived",axis=1)
y=train_df["Survived"]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)

# Here fitted the data and predicted for testing data
ypred_test=lr.predict(x_test)
ypred_test
```

```python
from sklearn.metrics import classification_report
import sklearn.metrics as metrics

# Here the accuracy b/w actual and prediction of testing data

print(classification_report(y_test,ypred_test))
print("Accuracy:",metrics.accuracy_score(y_test,ypred_test).round(2))

from sklearn.metrics import precision_score, recall_score

print("Precision:", precision_score(y_test,ypred_test).round(2))
print("Recall:", recall_score(y_test,ypred_test).round(2))
```

```python
import numpy as np

train_acc,test_acc=[],[]
for i in range(101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=i)
    lr.fit(x_train,y_train)
    train_acc.append(metrics.accuracy_score(y_train,lr.predict(x_train)))
    test_acc.append(metrics.accuracy_score(y_test,lr.predict(x_test)))
print(f"Train Accuracy :{np.mean(train_acc).round(2)}")
print(f"Test Accuracy :{np.mean(test_acc).round(2)}")
```

```python
from sklearn.metrics import roc_curve,roc_auc_score

train_df["y_proba"]=lr.predict_proba(x)[:,1]
fpr,tpr,dummy=roc_curve(y,train_df["y_proba"])
```

```python
plt.figure(figsize=(4,3))
plt.scatter(fpr,tpr)
plt.plot(fpr,tpr,color="red")
```

```python
print("AUC score:",roc_auc_score(y,train_df["y_proba"]).round(3))
```

```python
## What is the difference between precision and recall?

# Precision is Out of all the predictions my model predicted were positive, how many were actually positive.
# Recall is Out of all the actual positive predictions, how many did my model predicted correctly.
```
